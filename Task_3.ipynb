```python
# part 2 Data  Analaysis #  groupby()
import pandas as pd

# Load dataset
df = pd.read_csv("/content/train.csv")

# --- Step 2: Data Cleaning (essential for analysis) ---
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns='Cabin')
df = df.drop_duplicates()

# --- Step 3a: Survival rate by gender ---
survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("=== Survival Rate by Gender ===")
print(survival_by_gender)

# --- Step 3b: Survival rate by class (Pclass) ---
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("\n=== Survival Rate by Class ===")
print(survival_by_class)

# --- Step 3c: Average age per class ---
average_age_per_class = df.groupby('Pclass')['Age'].mean()
print("\n=== Average Age per Class ===")
print(average_age_per_class)

# Create age groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teen', 'Adult', 'Middle-Aged', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Use observed=True to avoid FutureWarning
survival_by_age_group = df.groupby('AgeGroup', observed=True)['Survived'].mean()
print("\n=== Survival Rate by Age Group ===")
print(survival_by_age_group)
```

```python
# data filter
import pandas as pd

# Load dataset
df = pd.read_csv("/content/train.csv")

# --- Data Cleaning ---
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns='Cabin')
df = df.drop_duplicates()

# --- Set Pandas display options for wide output ---
pd.set_option('display.width', 1000)        # wide display width
pd.set_option('display.max_columns', 10)    # show all columns
pd.set_option('display.colheader_justify', 'center')  # center headers

# --- Female passengers who survived ---
female_survived = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
print("=== Female Passengers Who Survived ===")
display(female_survived[['Name', 'Sex', 'Pclass', 'Survived']].head(10))

# --- Children who survived (Age < 12) ---
children_survived = df[(df['Age'] < 12) & (df['Survived'] == 1)]
print("\n=== Children Who Survived ===")
display(children_survived[['Name', 'Age', 'Pclass', 'Survived']].head(10))

# --- 1st class passengers with high survival probability ---
first_class_survived = df[(df['Pclass'] == 1) & (df['Survived'] == 1)]
print("\n=== 1st Class Passengers Who Survived ===")
display(first_class_survived[['Name', 'Pclass', 'Sex', 'Survived']].head(10))
```

```python
# ========================================
# Step 5: Insights - Titanic Dataset
# ========================================

import pandas as pd

# Load the dataset (assume it has already been cleaned)
df = pd.read_csv("/content/train.csv")

# Fill missing values (required for analysis)
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df = df.drop(columns='Cabin')
df = df.drop_duplicates()

# ------------------------
# Step 5: Insights
# ------------------------
print("=== Step 5: Insights ===")

# 1️⃣ Gender survival
gender_survival = df.groupby('Sex')['Survived'].mean()
print(f"1️⃣ Gender: Females were more likely to survive ({gender_survival['female']:.2f}) than males ({gender_survival['male']:.2f})")

# 2️⃣ Class survival
class_survival = df.groupby('Pclass')['Survived'].mean()
print(f"2️⃣ Class: 1st class ({class_survival[1]:.2f}) survived more than 2nd class ({class_survival[2]:.2f}) and 3rd class ({class_survival[3]:.2f})")

# 3️⃣ Children survival
children_rate = df[df['Age'] < 12]['Survived'].mean()
print(f"3️⃣ Children: Children (<12 years) had higher survival rate ({children_rate:.2f}) than adults")

# 4️⃣ Highest survival combination
highest_combo = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]['Survived'].mean()
print(f"4️⃣ Highest survival combination: Female passengers in 1st class had the highest survival probability ({highest_combo:.2f})")
```
